```python
#| export
from typing import Any

from fasthtml.common import Script

from cjm_fasthtml_viewport_fit.models import ViewportFitConfig
from cjm_fasthtml_viewport_fit.js import generate_viewport_fit_js
```

## Render Viewport Fit Script

Returns a FastHTML `Script` element containing the complete viewport-fit JS. Include this in the page alongside the target element.

```python
#| export
def render_viewport_fit_script(
    config: ViewportFitConfig,  # Instance configuration
) -> Any:  # Script element with complete viewport-fit JS
    """Render a Script element with the complete viewport-fit measurement JS."""
    return Script(generate_viewport_fit_js(config))
```

## Tests

```python
from fasthtml.common import to_xml

cfg = ViewportFitConfig(namespace="test", target_id="my-target")
script = render_viewport_fit_script(cfg)
html = to_xml(script)

# Verify it's a script tag
assert html.startswith("<script>")
assert html.endswith("</script>")

# Verify content includes key functions
assert "_calculateAndSetHeight" in html
assert "recalculateTestHeight" in html

# Namespace isolation: two configs produce different handler keys
cfg_a = ViewportFitConfig(namespace="alpha", target_id="a")
cfg_b = ViewportFitConfig(namespace="beta", target_id="b")
html_a = to_xml(render_viewport_fit_script(cfg_a))
html_b = to_xml(render_viewport_fit_script(cfg_b))
assert "_vfResizeHandler_alpha" in html_a
assert "_vfResizeHandler_beta" in html_b
assert "_vfResizeHandler_alpha" not in html_b

print("All tests passed!")
```
